```python
import pandas as pd
import matplotlib.pyplot as plt

# Set matplotlib parameters for IEEE paper standard
plt.rcParams["figure.figsize"] = (6.9, 4.5)  # Width: 6.9 inches (single column IEEE)
plt.rcParams["font.size"] = 10
plt.rcParams["axes.labelsize"] = 10
plt.rcParams["axes.titlesize"] = 10
plt.rcParams["xtick.labelsize"] = 8
plt.rcParams["ytick.labelsize"] = 8
plt.rcParams["legend.fontsize"] = 8
plt.rcParams["figure.dpi"] = 300
plt.rcParams["savefig.dpi"] = 300
plt.rcParams["savefig.format"] = "pdf"
plt.rcParams["lines.linewidth"] = 1
plt.rcParams["axes.grid"] = True
plt.rcParams["grid.linestyle"] = "--"
plt.rcParams["grid.alpha"] = 0.7
```

```python
column_names = [
    "unit_number",
    "time_cycles",
    "operational_setting_1",
    "operational_setting_2",
    "operational_setting_3",
    "sensor_measurement_1",
    "sensor_measurement_2",
    "sensor_measurement_3",
    "sensor_measurement_4",
    "sensor_measurement_5",
    "sensor_measurement_6",
    "sensor_measurement_7",
    "sensor_measurement_8",
    "sensor_measurement_9",
    "sensor_measurement_10",
    "sensor_measurement_11",
    "sensor_measurement_12",
    "sensor_measurement_13",
    "sensor_measurement_14",
    "sensor_measurement_15",
    "sensor_measurement_16",
    "sensor_measurement_17",
    "sensor_measurement_18",
    "sensor_measurement_19",
    "sensor_measurement_20",
    "sensor_measurement_21",
]

# Read the CSV file with the defined column names
df = pd.read_csv(
    "dados/6.turbofan rul/train_FD001.txt", sep=" ", header=None, names=column_names
)

# Drop the last two columns which are empty (NaN values)
# df = df.dropna(axis=1, how='all')
```

```python
import seaborn as sns

# Get all columns except unit_number
cols_to_plot = column_names[2:]  # Skip unit_number

# Looking at unique unit values
unit_values = df.unit_number.unique()

# Select the second unique unit value (since they're floats, not integers)
unit_value = unit_values[1]  # This is 0.0019


# Filter data for the selected unit
unit_data = df[df.unit_number == unit_value].reset_index(drop=True)

# Create a figure with subplots
fig, axes = plt.subplots(nrows=13, ncols=2, figsize=(12, 28))


axes = axes.flatten()

# Plot each measurement on its own subplot
for i, col in enumerate(cols_to_plot):
    if i < len(axes):  # Ensure we don't exceed the number of subplots
        ax = axes[i]

        # Use unit_data instead of df to only plot data for the selected unit
        sns.histplot(data=unit_data, x=col, ax=ax, kde=True)

        ax.set_title(col)

        ax.set_xlabel(col)  # Fix the f-string
        ax.set_ylabel("Frequency")

# Adjust layout
plt.tight_layout()

# Display the figure
plt.show()

# Print the shape of the filtered data
print(f"Shape of unit {unit_value} data: {unit_data.shape}")
```

```python
# Get all columns except unit_number
cols_to_plot = column_names[1:]  # Skip unit_number but keep time_cycles in the list

# Looking at unique unit values
unit_values = df.unit_number.unique()

# Select the second unique unit value
unit_value = unit_values[1]  # This is 0.0019

# Filter data for the selected unit
unit_data = df[df.unit_number == unit_value].reset_index(drop=True)

# Create a figure with subplots
fig, axes = plt.subplots(nrows=13, ncols=2, figsize=(12, 28))
axes = axes.flatten()

# Plot each measurement on its own subplot
for i, col in enumerate(cols_to_plot):
    if i < len(axes) and col != "time_cycles":  # Skip time_cycles as it's our x-axis
        ax = axes[i]

        # Plot the line chart with time_cycles on x-axis and sensor measurement on y-axis
        sns.lineplot(data=unit_data, x="time_cycles", y=col, ax=ax)

        ax.set_title(col)
        ax.set_xlabel("Time Cycles")
        ax.set_ylabel("Value")

# Adjust layout
plt.tight_layout()

# Display the figure
plt.show()

# Print the shape of the filtered data
print(f"Shape of unit {unit_value} data: {unit_data.shape}")
```
